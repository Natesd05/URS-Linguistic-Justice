```python
!pip install langid
import langid
import re
import pandas as pd

from langid.langid import LanguageIdentifier
from langid.langid import model

class LanguageIdentification:
    def __init__(self, file_path):
        self.spanish_words = 0
        self.english_words = 0
        self.total_words = 0
        self.non_counted = 0
        self.word_bank = ""
        with open(file_path, 'r') as file:
            self.word_bank = file.read()
        self.word_list = [token for token in re.split('[.?!]', self.word_bank) if token]
        self.lang_identifier = LanguageIdentifier.from_modelstring(model, norm_probs=True)

    def lang_detect(self, sentence):
        try:
            lang, prob = self.lang_identifier.classify(sentence)
            return lang, prob
        except:
            return None, None

    def perform_language_detection(self):
      results = []
      for str in self.word_list:
          self.total_words += 1
          lang, prob = self.lang_detect(str)
          if prob is not None and prob > 0.4:
            results.append({'sentence': str, 'language': lang, 'probability': prob})
            if (lang == ("es")):
              self.spanish_words += 1
            elif (lang ==("en")):
              self.english_words += 1
          else:
            self.non_counted += 1
      df = pd.DataFrame(results)
      return df

    def print_spanish(self):
      print(self.spanish_words)

    def print_english(self):
      print(self.english_words)

    def percentage_spanish(self):
      print(self.spanish_words/self.total_words)

    def print_total_words(self):
      print(self.total_words)

    def print_non_counted(self):
      print(self.non_counted)


language_identifier = LanguageIdentification("/TexasInterview3.txt")
result_df = language_identifier.perform_language_detection()

english_spanish_df = result_df[result_df['language'].isin(['en', 'es'])]
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

print(english_spanish_df)

pd.reset_option('display.max_rows')
pd.reset_option('display.max_columns')


language_identifier.print_spanish()
language_identifier.print_english()
language_identifier.percentage_spanish()
language_identifier.print_total_words()
language_identifier.print_non_counted()
```
